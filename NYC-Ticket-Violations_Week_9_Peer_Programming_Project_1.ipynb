```python
import pandas as pd

df = pd.read_csv('nyc-parking-violations-2020_.csv')
```

```python
# Drop rows with missing values one or more in the specified columns
# so we tried to use License plate as Plate ID, state as Registration State, and no need to use any other reference of words coz we have Street Name column.
# license plate, state, and/or street name


df_1 = df.dropna(subset=['Plate ID', 'Registration State', 'Street Name'])

# Print the number of rows in the filtered dataset
print('Number of rows remain in the filtered dataset are:', df_1.shape[0])
```

```python
# Let's Calculate the revenue from the original dataset before removing some rows
revenue_original = df.shape[0] * 100

# Let's Calculate the revenue from the filtered dataset after removing some rows
revenue_filtered = df_1.shape[0] * 100

# Let's Calculate the revenue lost by the City as a result of this
revenue_lost = revenue_original - revenue_filtered

# Now let me diplay the moeny lost by the city
print('Revenue lost by city is : $', revenue_lost)
```
